This project aims to predict the personality type (Extrovert vs Introvert) of a person based on key characteristics. By analysing patterns between different data types, I aim to explore how specific behavioural traits correleate with one's personality type. 

```python
import pandas as pd
df = pd.read_csv('personality_dataset.csv')
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df[['Time_spent_Alone', 'Social_event_attendance','Going_outside', 'Friends_circle_size','Post_frequency']].mean()
```

```python
df[['Time_spent_Alone', 'Social_event_attendance','Going_outside', 'Friends_circle_size','Post_frequency']] = df[['Time_spent_Alone', 'Social_event_attendance','Going_outside', 'Friends_circle_size','Post_frequency']].fillna({
    'Time_spent_Alone' : 4.505816 ,
    'Social_event_attendance' : 3.963354,
    'Going_outside' : 3.000000,
    'Friends_circle_size' : 6.268863,
    'Post_frequency' : 3.564727
})
df.info()
```

```python
df[['Stage_fear', 'Drained_after_socializing']].mode()
```

```python
df[['Stage_fear', 'Drained_after_socializing']] = df[['Stage_fear', 'Drained_after_socializing']].fillna('No')
df.info()
```

```python
df['Stage_fear'] = df['Stage_fear'].astype(str).str.strip().str.lower()
df['Drained_after_socializing'] = df['Drained_after_socializing'].astype(str).str.strip().str.lower()
```

```python
df['Stage_fear'] = df['Stage_fear'].map({'no' : 0.0, 'yes' : 1.0})
df['Drained_after_socializing'] = df['Drained_after_socializing'].map({'no' : 0.0, 'yes' : 1.0})
df['Personality'] = df['Personality'].map({
    'Introvert' : 0.0, 
    'Extrovert' : 1.0
})
df.info()
```

```python
df['Drained_after_socializing'].head(10)
```

```python
x = df[['Time_spent_Alone', 'Stage_fear', 'Social_event_attendance', 'Going_outside', 'Drained_after_socializing', 'Friends_circle_size', 'Post_frequency']]
y = df['Personality']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

I split the data using the 80/20 rule instead of the 70/30 rule because the dataset I am working with is large and a better split would be 80/20. 

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree_model = DecisionTreeClassifier(random_state = 0)
decision_tree_model.fit(X_train, Y_train)
```

In this the decision tree model was created, while keeping the random_state = 0, to ensure that the result stays the same. 

```python
Y_predic = decision_tree_model.predict(X_test)
Y_predic
```

```python
from sklearn.metrics import confusion_matrix
conmat_pred = confusion_matrix(Y_test, Y_predic)
conmat_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(Y_test, Y_predic)
accuracy
```

```python
precision = precision_score(Y_test, Y_predic)
precision
```

```python
recall = recall_score(Y_test, Y_predic)
recall
```

```python
f1 = f1_score(Y_test, Y_predic)
f1
```

Interpretation: The model correctly identified 273 Introverts and 234 Extroverts, while misidentifying 37 introverts as extroverts and 36 extroverts as introverts. In conclusion, out of 580 test samples, 507 were correctly identified, giving the model an accuracy of 87.4%. Apart from this, the model has a precision rate of 86.3%, recall rate of 86.7% and F1 score of 86.5%, illustrating the effective prediction of personality type, with a minimal misclassification rate and fairly equal performance in both classes. 


```python
from sklearn.metrics import classification_report
class_rep = classification_report(Y_test, Y_predic)
class_rep
```

Conclusion

After performing comprehensive data preprocessing and feature encoding, a Decision Tree Classifier was developed to categorize personality types into groups. The model demonstrated strong classification performance, achieveing an overall accuracy of 87.4% on the test dataset.
